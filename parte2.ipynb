```python
import pandas as pd
```

```python
"""
Parte 2 : No Jupyter vocês poderão importar ou criar uma tabela do excel com pelo menos 5 colunas como um dataframe.  
Quebrar essa tabela em duas e após fazer merge do tipo LEFT JOIN. (Aula03) (Fizemos exercícios em sala demonstrando esses dois passos, 
vocês podem replicar, com uma tabela diferente). Nos links do teams, em cada aula tem a sintaxe e a explicação detalhadamente.
"""

# Importa tabela do arquivo xlsx
df_principal: pd.DataFrame = pd.read_excel("tabela.xlsx")
df_principal
```

```python
# Extrai para uma nova tabela para informacoes fornecedor (CNPJ e Nome do fornecedor)
# E remove dados duplicados
df_fornecedores = pd.DataFrame({
    "CNPJ do Fornecedor": df_principal["CNPJ do Fornecedor"],
    "Nome do Fornecedor": df_principal["Nome do Fornecedor"]
}).drop_duplicates()
df_fornecedores
```

```python
# Tira informacoes do fornecedor da tabela principal
# O axis significa que queremos remover da tabela pela coluna
df_principal_sem_fornecedor = df_principal.drop([ "Nome do Fornecedor"], axis=1)
df_principal_sem_fornecedor
```

```python
# Liga tabela sem fornecedor com a de informacoes de fornecedor via coluna "CNPJ do Fornecedor"
df_join = df_principal_sem_fornecedor.join(df_fornecedores.set_index("CNPJ do Fornecedor"), on="CNPJ do Fornecedor", how="left").drop_duplicates()
df_join
```
