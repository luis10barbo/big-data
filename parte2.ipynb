```python
"""
Parte 2 : No Jupyter vocÃªs poderÃ£o importar ou criar uma tabela do excel com pelo menos 5 colunas como um dataframe.  
Quebrar essa tabela em duas e apÃ³s fazer merge do tipo LEFT JOIN. (Aula03) (Fizemos exercÃ­cios em sala demonstrando esses dois passos, 
vocÃªs podem replicar, com uma tabela diferente). Nos links do teams, em cada aula tem a sintaxe e a explicaÃ§Ã£oÂ detalhadamente.
"""
df_principal: pd.DataFrame = pd.DataFrame()
with open("tabela.xlsx", "rb") as arquivo_tabela:
    df_principal = pd.read_excel(arquivo_tabela)

df_principal
```

```python

dfFornecedores = pd.DataFrame({
    "CNPJ do Fornecedor": df_principal["CNPJ do Fornecedor"],
    "Nome do Fornecedor": df_principal["Nome do Fornecedor"]
}).drop_duplicates()

df_sem_fornecedor = df_principal.drop([ "Nome do Fornecedor"], axis=1)
```

```python
df_join = df_sem_fornecedor.join(dfFornecedores.set_index("CNPJ do Fornecedor"), on="CNPJ do Fornecedor", how="left").drop_duplicates()
df_join
```
